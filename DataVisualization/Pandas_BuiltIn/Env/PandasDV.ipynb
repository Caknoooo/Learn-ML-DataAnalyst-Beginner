## Pandas Built-in Data Visualization
In this lecture we will learn about pandas built-in capabilities for data visualization

```python
import numpy as np 
import pandas as pd
%matplotlib inline
```

```python
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']},index=[0, 1, 2, 3])

df2 = pd.DataFrame({
    'A': ['A4', 'A5', 'A6', 'A7'],
    'B': ['B4', 'B5', 'B6', 'B7'],
    'C': ['C4', 'C5', 'C6', 'C7'],
    'D': ['D4', 'D5', 'D6', 'D7']},index=[4, 5, 6, 7]) 

df2
```

### Style sheets
Matplotlib has style sheets you can use to make your plots look a little nicer. These style sheets include plot_bmh and more. They basically create a set of style rules that your plots follow. I recommended using them, they make all your plots have the same look and feel more proffesional. You can even create you own if you want your company's plots to all have the same look

```python
df1['A'].hist()
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
df1['A'].hist()
```

```python
plt.style.use('bmh')
df1['A'].hist()
```

```python
plt.style.use('dark_background')
df1['A'].hist()
```
