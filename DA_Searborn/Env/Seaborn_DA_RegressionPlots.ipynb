## Regression Plots
Seaborn has many built0in capabilities for regression plots, however we won't really discuss regression until the machine learning section of the course, so we will only cover the lmplot() function for now.

Implot allows you to display linear models, but it also conveniently allows you yo splitup those plots based off of  features, as well as coloring the hue based off of features.

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

### lmplot()

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='rainbow')
```

### Working with Markers
lmplot kwargsget passed through to rgplot which is a more general form of lmplot(). regplot has a scatter _kws parameter that gets passed to plt.scatter(). So you want to set the parameter in that dictionary, which corresponds (a bit confusingly) to the squared markersize. In other words you end up passing a dictionary with the base matplotlib arguments, in this case, s for size of a scatter plot. In general, you probably won't remember this off the top of yout head, but instead  reference the documentation.

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='coolwarm'
    , markers=['o', 'v'], scatter_kws={'s':100})
```

### Using a Grid
We can add more variable separation through columns and rows with the use of a grid. Just indicate this with the col or row arguments

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', row='sex', col='time', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='sex', palette='coolwarm')
```

### Ascept and Size
Seaborn figures can have their size and aspect ratio adjusted with the size and aspect parameters

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm',
          aspect=0.6)
```
